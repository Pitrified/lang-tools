The learner reorders shuffled portions of a target-language sentence.

```python
from loguru import logger as lg
from rich import get_console
from rich import print as rprint
from rich.console import Console

# some magic to make rich work in jupyter
# https://github.com/Textualize/rich/issues/3483
# enable it for every cell output with %load_ext rich
console: Console = get_console()
console.is_jupyter = False
```

```python
import random

from lang_tools.exercises import SentenceReconstructionExercise
from lang_tools.exercises.sentence_reconstruction import merge_short_portions
```

## Merge short portions


```python
rprint(merge_short_portions(["Eu", "gosto", "de", "café"]))
rprint(merge_short_portions(["I", "love", "coffee"]))
```

## Run a round


```python
exercise = SentenceReconstructionExercise(rng=random.Random(0))
round_ = exercise.start(
    sentence="Eu gosto muito de café",
    translation="I really like coffee",
)
rprint(round_.prompt)
rprint("expected:", round_.expected)
```

```python
result = exercise.submit(round_, round_.expected)
rprint(result)
```

```python
rprint(exercise.finish())
```
