# Pair matching exercise

Use `PairMatchingExercise` to match words to their translations.

```python
%load_ext autoreload
%autoreload 2
```

```python
import random

from rich import print as rprint

from lang_tools.exercises import PairMatchingExercise
from lang_tools.words.word import Word
```

## Build a small vocabulary

```python
words = [
    Word(
        text="amor",
        language="pt",
        part_of_speech="noun",
        frequency="high",
        translations={"en": "love"},
    ),
    Word(
        text="paz",
        language="pt",
        part_of_speech="noun",
        frequency="high",
        translations={"en": "peace"},
    ),
    Word(
        text="casa",
        language="pt",
        part_of_speech="noun",
        frequency="high",
        translations={"en": "house"},
    ),
]
```

## Run a round

```python
exercise = PairMatchingExercise(target_language="en", rng=random.Random(0))
round_ = exercise.start(words)
rprint(round_.prompt)
```

```python
result = exercise.submit(round_, ("amor", "love"))
rprint(result)
result = exercise.submit(round_, ("paz", "house"))
rprint(result)
```

```python
rprint(exercise.finish())
```
