# Diacritic typing exercise

Guide the user to type accented characters one keystroke at a time.

```python
%load_ext autoreload
%autoreload 2
```

```python
from rich import print as rprint

from lang_tools.exercises import DiacriticTypingExercise
from lang_tools.words.word import Word
```

```python
word = Word(text="Ação", language="pt", part_of_speech="noun", frequency="high")
exercise = DiacriticTypingExercise()
round_ = exercise.start(word, hint_level="show_unaccented")
rprint(round_.prompt)
```

```python
for ch in "Ação":
    result = exercise.submit(round_, ch)
    rprint(ch, result.correct, round_.prompt["display"])
```

```python
rprint(exercise.finish())
```
