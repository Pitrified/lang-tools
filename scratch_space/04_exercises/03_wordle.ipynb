# Wordle exercise

Guess the hidden word; each letter is graded `correct` / `misplaced` / `wrong`.

```python
%load_ext autoreload
%autoreload 2
```

```python
from rich import print as rprint

from lang_tools.exercises import WordleExercise
from lang_tools.words.word import Word
```

```python
target = Word(text="amor", language="pt", part_of_speech="noun", frequency="high")
exercise = WordleExercise()
round_ = exercise.start(target)
rprint(
    "word_length:",
    round_.prompt["word_length"],
    "max_attempts:",
    round_.prompt["max_attempts"],
)
```

```python
for guess in ["casa", "amor"]:
    result = exercise.submit(round_, guess)
    rprint(
        guess,
        result.correct,
        [(lr.letter, lr.state) for lr in round_.prompt["results"][-1]],
    )
```

```python
rprint(exercise.finish())
```
