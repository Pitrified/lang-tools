# Word normalisation and the `Word` model

Demonstrates the `lang_tools.language` and `lang_tools.words` packages.


```python
%load_ext autoreload
%autoreload 2
```

```python
from rich import print as rprint

from lang_tools.language import LANGUAGE_PRESETS
from lang_tools.language import extract_accented_chars
from lang_tools.language import get_language
from lang_tools.language import has_accent
from lang_tools.language import normalize
from lang_tools.words.word import Word
```

## Language presets


```python
rprint(sorted(LANGUAGE_PRESETS.keys()))
pt = get_language("pt")
rprint(pt.code, pt.name, sorted(pt.accented_chars))
```

## Normalising text


```python
samples = ["Café", "Ação", "Nação", "MÃOS"]
for s in samples:
    rprint(s, "->", normalize(s, pt), "accents:", extract_accented_chars(s))
rprint("has_accent('cafe'):", has_accent("cafe"))
rprint("has_accent('café'):", has_accent("café"))
```

## Building a `Word`


```python
w = Word(
    text="Ação",
    language="pt",
    part_of_speech="noun",
    frequency="high",
    translations={"en": "action"},
    topics=["abstract"],
)
rprint(w)
rprint("id:", w.id, "normalized:", w.normalized, "len:", w.length, "accented:", w.accented_chars)
```
