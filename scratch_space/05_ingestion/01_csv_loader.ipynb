# CSV ingestion

Load a small vocabulary file into `Word` instances using `load_csv`.

```python
%load_ext autoreload
%autoreload 2
```

```python
from io import StringIO

from rich import print as rprint

from lang_tools.words.ingestion import load_csv
```

```python
content = (
    "text,language,part_of_speech,frequency,translation_en,topics\n"
    "amor,pt,noun,high,love,emotions\n"
    "paz,pt,noun,high,peace,emotions\n"
    "café,pt,noun,medium,coffee,food\n"
)
for w in load_csv(StringIO(content)):
    rprint(w)
```
